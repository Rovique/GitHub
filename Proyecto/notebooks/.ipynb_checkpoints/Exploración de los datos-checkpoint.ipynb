## 4.3 Ingenier√≠a de variables

```python
import datetime as dt
```

```python
cero = dt.datetime.now()
cero
```

```python
in_time_aux = in_time[in_time.columns[1::]].copy()
out_time_aux = out_time[out_time.columns[1::]].copy()

in_time_aux.fillna(cero)
out_time_aux.fillna(cero)
```

```python
aux = pd.DataFrame()

for c in in_time.columns[1::]:
    aux[c] = pd.to_datetime(out_time_aux[c]) - pd.to_datetime(in_time_aux[c])

aux
```

```python
holidays = pd.DataFrame(aux.describe())
holidays
```

```python
holidays = holidays.T.copy()
holidays = list(holidays[holidays.loc[:,'count'] == 0].index)

aux = aux[[x for x in aux.columns if x not in holidays]].copy()

totaldays = len(aux.T)
totalhours = aux.T.sum()
minimum = aux.T.min()
maximum = aux.T.max()
mean = aux.T.mean()
std = aux.T.std()
absences = aux.T.count()
absences = [totaldays - x for x in absences]
absences_pct = [x/totaldays for x in absences]
of_absences = len(holidays)

dt_columns = ['total_hours','minimum_hours', 'maximum_hours', 'mean_hours', 'std_min', 'absences', 'absences_pct', 'of_absence']
dt_values = [totalhours,minimum, maximum, mean, std, absences, absences_pct, of_absences]
aux_dict = dict(zip(dt_columns, dt_values))

dt_df = pd.DataFrame()

dt_df['EmployeeID'] = in_time['Unnamed: 0']

for c in dt_columns:
    dt_df[c] = aux_dict[c]
    
dt_df['total_absences'] = dt_df['absences'] + dt_df['of_absence']
dt_df['total_hours'] = dt_df['total_hours'].map(lambda x:x.days*24)
dt_df['Expected_time_vs_real_time'] = [1 for i in range(len(dt_df))]
dt_df['minimum_hours'] = dt_df['minimum_hours'].map(lambda x:x.seconds/3600)
dt_df['maximum_hours'] = dt_df['maximum_hours'].map(lambda x:x.seconds/3600)
dt_df['mean_hours'] = dt_df['mean_hours'].map(lambda x:x.seconds/3600)
dt_df['std_min'] = dt_df['std_min'].map(lambda x:x.seconds/60)
dt_df['Expected_time_vs_real_time'] = dt_df['total_hours']/((261-dt_df['of_absence'])*8)
dt_df['Expected_vacations_time_vs_real_time'] = dt_df['total_hours']/((261-dt_df['total_absences'])*8)
```

# 5. Cruce de datos

En este espacio, juntaremos las bases, para nutrir nuestra tabla general a nivel empleado, para esto utilizaremos las siguientes tablas:

* general
* employee_survey
* manager_survey
* dt_df

Y en todas la llave es 'EmployeeID'.

```python
df_final = general.merge(employee_survey, how = 'inner', left_on='EmployeeID', right_on='EmployeeID')
df_final = df_final.merge(manager_survey, how = 'inner', left_on='EmployeeID', right_on='EmployeeID')
df_final = df_final.merge(dt_df, how = 'inner', left_on='EmployeeID', right_on='EmployeeID')
df_final.head(5)
```
